```python
%matplotlib inline
from __future__ import print_function

import sys
import numpy as np
import math

import first
import thinkplot
import thinkstats2
```

```python
live, firsts, others = first.MakeFrames()
live = live.dropna(subset=['agepreg', 'totalwgt_lb'])
ages = live.agepreg
weights = live.totalwgt_lb
```

```python
print('thinkstats2 Corr', thinkstats2.Corr(ages, weights))
print('thinkstats2 SpearmanCorr', 
          thinkstats2.SpearmanCorr(ages, weights))
```

```python
thinkplot.Scatter(ages, weights)
thinkplot.Config(xlabel='age (years)',
                     ylabel='weight (lbs)',
                     legend=False)
thinkplot.Show()
```

There doesn't seem to be a relationship between the two. This is also backed up by the Pearson's correlation and Spearman's rank correlation.
